```python
import math
from tqdm import tqdm
import matplotlib.pyplot as plt
```

# Plot JL Lemma 

$$\text{min compoment} \geq \frac{4*\log{n_{samples}}}{\epsilon^2/2 - \epsilon^3/3}$$ 

```python

eps = [0.5,0.6,0.7,0.8,0.9,1.0]
n_samples_lst = [x for x in range(100,100000, 100)]
n_component_lst = {}

for e in tqdm(eps):
    n_component_lst[e] = []
    for n_samples in n_samples_lst :
        min = int(math.ceil(4* math.log(n_samples) / (e**2/2 - e**3/3))) 
        n_component_lst[e].append(min)
    
for e in eps:
    plt.plot(n_samples_lst, n_component_lst[e], label="eps="+str(e))
plt.xlabel("n_samples", fontsize=12)
plt.ylabel("min_components", fontsize=12)
plt.title("JL", fontsize=14)
plt.legend(fontsize=12)
plt.show()
```
